```python
### Get historical prices data from Stooq ###
import pandas_datareader.data as web
df = web.DataReader('AAPL', 'stooq')
df.tail(10)
```

```python
### Get historical prices data from Yahoo! Finance ###
import pandas_datareader.data as web
df = web.DataReader('AAPL', 'yahoo')
df.tail(10)
```

```python
### Get historical prices data from Yahoo! Finance (other way) ###
import pandas_datareader.data as web
df = web.get_data_yahoo('AAPL', '2020-05-01', '2020-07-02')
df.tail(10)
```

```python
### Get historical prices from Quandl ###
import quandl
symbol = 'WIKI/AAPL'
df = quandl.get(symbol, start_date='2018-01-01', end_date='2020-07-02')
df.tail(10)
```

```python
### Get historical prices Quandl (only get specific columns) ###
import quandl
df = quandl.get(["WIKI/FB.1", "WIKI/AAPL.4"])
df.tail(10)
```

```python
### Get historical prices from Quandl (only latest rows) ###
import quandl
df = quandl.get("WIKI/AAPL", rows=5)
df.tail(10)
```
